# Accuracy Plot

We now compare the accuracy of our models for different input dimensions (5, 10, 20, 50, 100).

```python
import numpy as np
from collections import defaultdict
%matplotlib nbagg
import random
import matplotlib.pyplot as plt
import dynet
import torch
import torch.nn.functional as F
```

```python
def evaluate_test(w_test_file,data_file):
    true=0
    compare=open(data_file,'r')
    l=[]
    k=[]
    for sentence1 in compare:
        words1=sentence1.strip().split()
        if len(words1)==2:
            l.append(words1[1])
    for sentence2 in open(w_test_file,'r'):
        words2=sentence2.strip().split()
        if len(words2)==2:
            k.append(words2[1])
    for i in range(len(l)):
        if l[i]==k[i]:
            true+=1
    accuracy=true/len(l)
    return accuracy
```

```python
eucl2=[]
hyper=[]
base=[]
```

```python
eucl2.append(evaluate_test('data/en.pos.dev.raw.output.eucl.dim5.square','data/en.pos.dev'))
eucl2.append(evaluate_test('data/en.pos.dev.raw.output.eucl.dim10.square','data/en.pos.dev'))
eucl2.append(evaluate_test('data/en.pos.dev.raw.output.eucl.dim20.square','data/en.pos.dev'))
eucl2.append(evaluate_test('data/en.pos.dev.raw.output.eucl.dim50.square','data/en.pos.dev'))
eucl2.append(evaluate_test('data/en.pos.dev.raw.output.eucl.dim100.square','data/en.pos.dev'))
```

```python
hyper.append(evaluate_test('data/en.pos.dev.raw.output.hyperbballbias.dim5','data/en.pos.dev'))
hyper.append(evaluate_test('data/en.pos.dev.raw.output.hyperbballbias.dim10','data/en.pos.dev'))
hyper.append(evaluate_test('data/en.pos.dev.raw.output.hyperbballbias.dim20','data/en.pos.dev'))
hyper.append(evaluate_test('data/en.pos.dev.raw.output.hyperbballbias.dim50','data/en.pos.dev'))
hyper.append(evaluate_test('data/en.pos.dev.raw.output.hyperbballbias.dim100','data/en.pos.dev'))
```

```python
base.append(evaluate_test('data/en.pos.dev.raw.output.pyex.dim5','data/en.pos.dev'))
base.append(evaluate_test('data/en.pos.dev.raw.output.pyex.dim10','data/en.pos.dev'))
base.append(evaluate_test('data/en.pos.dev.raw.output.pyex.dim20','data/en.pos.dev'))
base.append(evaluate_test('data/en.pos.dev.raw.output.pyex.dim50','data/en.pos.dev'))
base.append(evaluate_test('data/en.pos.dev.raw.output.pyex.dim100','data/en.pos.dev'))
```

```python
xaxis=['5','10','20','50','100']
random=[1/18,1/18,1/18,1/18,1/18]
plt.xlabel('Embedding Dimension')
plt.ylabel('Accuracy')
plt.plot(xaxis, base, 'r', label='Linear')
plt.legend()
plt.plot(xaxis, hyper, 'b', label='Hyper') 
plt.legend()
plt.plot(xaxis,eucl2,'y',label='Eucl')
plt.legend()
plt.plot(xaxis,random,'k',label='Random')
plt.legend()
plt.xticks(xaxis)
plt.show()
plt.savefig('accuracy.png')
```
